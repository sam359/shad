```python
d = {'one' : pd.Series([1., 2., 3.], index=['a', 'b', 'c']),'two' : pd.Series([1., 2., 3., 4.], index=['a', 'b', 'c', 'd'])}
pd.DataFrame(d)
```

```python
my_df = pd.DataFrame(d).unstack()
my_df.plot(kind='bar',stacked=True,title="My One/Two",colors=['red','blue'])
```
